```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
```

```python
def load_sheet(sheet_index, column_name):
    df = pd.read_excel("weather_data.xlsx", sheet_name=sheet_index, skiprows=3)
    df = df.dropna()
    df.columns = ['year', 'month'] + [str(i) for i in range(1, 32)]
    df = df.melt(id_vars=['year', 'month'], var_name='day', value_name=column_name)
    df['date'] = pd.to_datetime(df[['year', 'month', 'day']])
    df = df.sort_values(by='date')
    return df[['date', column_name]]


df0 = load_sheet(1, 'temp_avg')
df1 = load_sheet(2, 'temp_max')
df2 = load_sheet(3, 'temp_min')
df3 = load_sheet(4, 'wind_speed')
df4 = load_sheet(5, 'pressure')
df5 = load_sheet(6, 'humidity')
df6 = load_sheet(7, 'rain_cumulative')
df7 = load_sheet(8, 'snow_height')
df8 = load_sheet(9, 'sun')

df_weather = df0
for df in [df1, df2, df3, df4, df5, df6, df7, df8]:
    df_weather = df_weather.merge(df, how='outer', on='date')
```
